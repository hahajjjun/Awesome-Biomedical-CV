### What is Segmentation Task?

**Three main CV Tasks**
- Image Classification is the task of 'predicting labels'.
- Object Detection is the task of 'predicting bounding box'.
- Segmentation is the task of 'predicting pixel-wise labels'.

When there are multiple objects in the image, <br/>
**labeling every pixel to a label is important rather than bounding box**.

### Algorithms and approaches for segmentation

**01. FCN**
- E2E convolutional network
- Removing fully connected layers, replaced by convolution layer
- Also used for object detection task
- FCN is constructed without **maxpooling**
    - The important concept of segmentation is *preservation of spatial demension.*
- 1D convolution can be performed to reach desired number of labels

- Loss for this network is computed by **averaging the cross-entropy loss of every pixel & mini-batch**.

$$ \mathcal{L} = {{\sum_{i \in O} {\mathcal{L}_{ce}(p_i, b_i)}} \over {H \times W}}   \textrm{(Where } O \textrm{ is H} \times \textrm {W sized set of every pixels in output.)} $$
- Final layer has a depth equal to number of classes.
- Each pixel across the layer indicates the probability distribution of each pixel, hence ```argmax()``` will return labled image.

**02. SegNet**
- Encoder and Decoder approach.
- Encoder has various CONV layers **(Downsampling)** and Decoder **(Upsampling)** has various DECONV layers.
- SegNet improved the **coarse** outputs of FCN. (also call it **sparse**)
- Less intensive on memory.
